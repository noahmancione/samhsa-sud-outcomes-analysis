```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/tedsd_clean.csv')

print(df.shape)
df.head()
```

```python
# Completion rate by primary substance
substance_completion = (
    df.groupby('SUB1')['completed']
    .agg(['mean', 'count'])
    .rename(columns={'mean': 'completion_rate', 'count': 'total_episodes'})
    .sort_values('completion_rate', ascending=False)
)

substance_completion['completion_rate'] = substance_completion['completion_rate'].round(3)

print(substance_completion)
```

Opioid-related substances (heroin, other opiates, methadone) show the lowest completion rates around 14-24% while alcohol is at the top near 50%. Methamphetamine is a huge concern because of it being the second most-common substance while demonstrating a below average completino rate of 34%.

```python
# Completion rate by gender
gender_completion = (
    df.groupby('SEX')['completed']
    .agg(['mean', 'count'])
    .rename(columns={'mean': 'completion_rate', 'count': 'total_episodes'})
)

gender_completion['completion_rate'] = gender_completion['completion_rate'].round(3)
print(gender_completion)
```

Males complete at 43.2% and females at 41.5%. This gap seems small, but across nearly 1.5 million records, it presents a big gap. The gap likely reflects unique challenges facing women in treatment rather than differences in motivation or engagement.

```python
# Completion rate by age group
age_order = ['12-14', '15-17', '18-20', '21-24', '25-29', 
             '30-34', '35-39', '40-44', '45-49', '50-54', '55-64', '65+']

age_completion = (
    df.groupby('AGE')['completed']
    .agg(['mean', 'count'])
    .rename(columns={'mean': 'completion_rate', 'count': 'total_episodes'})
    .reindex(age_order)
)

age_completion['completion_rate'] = age_completion['completion_rate'].round(3)
print(age_completion)
```

The pattern is U-shaped. Adolescents (12-14) have the lowest completion rate at 35%. Middle aged patients cluster around 40-42%. Older adults (65+) have the highest rate at 56%. Older patients likely benefit from stronger self-motivation and more stable life circumstances.

```python
# Average length of stay by completion status
los_analysis = (
    df.groupby('completed')['LOS']
    .agg(['mean', 'median', 'count'])
    .round(1)
)
los_analysis.index = ['Not Completed', 'Completed']
print(los_analysis)
```

Completers have a median stay of 28 days versus 17 days for non-completers. The median is more meaningful than the mean here because a few very long stays skew the average. The 11-day gap suggests that early dropout in the first two weeks is the biggest driver of non-completion.

```python
# df is already loaded, create df_clean reference
df_clean = df.copy()

# Export analysis results for Power BI
substance_completion.to_csv('../data/substance_completion.csv')
gender_completion.to_csv('../data/gender_completion.csv')
age_completion.to_csv('../data/age_completion.csv')
los_analysis.to_csv('../data/los_analysis.csv')
df_clean.to_csv('../data/tedsd_clean.csv', index=False)

print("All files exported successfully")
```
